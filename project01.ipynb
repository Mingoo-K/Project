# 사각형 넓이 구하기

```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))
        print()

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
            return self.rect()

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
            return self.par()

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
            return self.trape()

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로, 세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        print()
        print(result)

    def par(self):
        width, base = map(int, input('높이, 밑변을 입력하세요 예시 : 높이, 밑변\n >>>').split(','))
        area = width * base
        result = '평행사변형의 넓이는 : ' + str(area)
        print()
        print(result)

    def trape(self):
        base1, base2, width = map(int, input('밑변1, 밑변2, 높이를 입력하세요 예시 : 밑변1, 밑변2, 높이\n >>>').split(','))
        area = (base1 + base2) * width / 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        print()
        print(result)

a = Square()
```

# 키오스크 만들기

```python
# 커피 메뉴와 가격
coffee = ['americano', 'latte', 'mocha']
c_price = [2000, 3000, 3000]

# 논커피 메뉴와 가격
non_coffee = ['yuza_tea', 'green_tea', 'choco_latte']
n_price = [2500, 2500, 3000]

menu = coffee + non_coffee  # 전체 음료수 메뉴
price = c_price + n_price   # 전체 음료수 가격

class Kiosk:
    def __init__(self):
        self.menu = menu
        self.price = price
        self.is_first_order = True  # 첫 주문 여부


    def menu_print(self): # 메뉴 출력
        for i in range(len(self.menu)):  # len(self.menu) = 6 = [0,1,2,3,4,5]
            print(i + 1, '. ', self.menu[i], ':', self.price[i], '원')
        print()

    def menu_select(self):  # 메뉴 선택
        n = 0
        while n < 1 or n > len(self.menu):
            n = int(input('음료의 번호를 입력하세요 : '))
            if 1 <= n <= len(self.menu):
              self.order_menu = []     # 주문 리스트
              self.order_price = []    # 가격 리스트
              self.menu_sum = self.menu[n - 1]  # 주문 번호 - 해당 메뉴 인덱스
              self.price_sum = self.price[n - 1]
              self.temp_select()
              self.order_menu.append(self.temp + ' ' + self.menu_sum) # 선택한 메뉴와 선택한 온도를 합침
              self.order_price.append(self.price_sum)
              if self.is_first_order:
                print('[ 주문 음료 - ', self.temp, self.menu_sum, ' : ', self.price_sum, '원 ]')
                self.is_first_order = False  # 첫 주문이 완료되었면 넘어감
            else:
                print('없는 메뉴입니다. 다시 주문해 주세요.\n')

    def temp_select(self):  # 음료 온도 물어보기
        t = 0
        while t != 1 and t != 2:     # 1이나 2를 입력할 때까지 물어보기
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:
                self.temp = 'HOT'
            elif t == 2:
                self.temp = 'ICE'
            else:
                print("1과 2 중 하나를 입력하세요.\n")

    def order_add(self):  # 추가 주문
        n = 1
        while n != 0:  # 지불을 선택하기 전까지 반복합니다.
            print()    # 줄 바꾸면서
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))
            if n > 0 and n <= len(self.menu):
                self.menu_sum = self.menu[n - 1]
                self.price_sum = self.price[n - 1]
                self.temp_select()
                self.order_menu.append(self.temp + ' ' + self.menu[n - 1])
                self.order_price.append(self.price_sum)
                print('[ 추가 주문 음료 : ', self.order_menu[-1], '-', self.order_price[-1], '원 ]')
                print('[ 현재 주문 합계 :', sum(self.order_price), '원 ]')
            elif n == 0:
                print('주문이 완료되었습니다.')
                print('최종 주문 내역 :', ', '.join(self.order_menu))
                print('최종 합계 금액 :', sum(self.order_price), '원')
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    def pay(self):
        print()
        p = 0
        while p != 1 and p != 2:
            p = int(input('현금 결제는 1번, 카드 결제는 2번을 입력하세요 : '))
            if p == 1:
                self.table()
                print('직원을 호출하겠습니다.')
            elif p == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요')
                self.table()
            else:
                print('다시 결제를 시도해주세요.')

    def table(self):
        print()
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        print()
        print('[ 합계 금액 :', sum(self.order_price), ']')

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')

# 테스트 코드
a = Kiosk()
a.menu_print()
a.menu_select()
a.order_add()
a.pay()
```
